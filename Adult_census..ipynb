```python
def selec_best_model(model_score):
    ms= list(model_score.values())
    ms_name= list(model_score.keys())
    best_score = max(ms)
    position = ms.index(best_score)
    
    return ms_name[position]
    
```

```python
best_model = selec_best_model(model_score)
```
